# Data Preparation and Model Discovery

```python
df.isnull().sum()
```

```python
df[df['Agriculture'].isnull()]['Country']
```

```python
#most of the above countries are small islands. hence we replace them by 0
df[df["Agriculture"].isnull()] = df[df["Agriculture"].isnull()].fillna(0)
```

```python
df[df['Agriculture'].isnull()].sum()
```

```python
df.isnull().sum()
```

```python
# we fill null values in climate by filling aerage climate values of the region in which the country lie.
df['Climate'] = df['Climate'].fillna(df.groupby("Region")['Climate'].transform('mean'))
```

```python
df.isnull().sum()
```

```python
# we fill null values in literacy column by putting the mean literacy values of the region.
df['Literacy (%)'] = df['Literacy (%)'].fillna(df.groupby('Region')['Literacy (%)'].transform('mean'))
```

```python
df.isnull().sum()
```

```python
# we are missing only few values. So we drop these null values
df = df.dropna()
```

```python
df.isnull().sum()
```

# Data Feature Preparation

```python
df.head()
```

```python
# we drop country column since it is a unique string identifier
X = df.drop('Country', axis=1)
```

```python
X.head()
```

```python
# creating dummy variables
X = pd.get_dummies(X)
```

```python
X.head()
```

# Scaling

```python
from sklearn.preprocessing import StandardScaler
```

```python
scaler = StandardScaler()
```

```python
scaled_X = scaler.fit_transform(X)
```

```python
scaled_X
```

# Creating and Fitting Kmeans Model

```python
from sklearn.cluster import KMeans
```

```python
ssd = []

for k in range(2,30):
    
    model = KMeans(n_clusters=k)
    
    model.fit(scaled_X)
    
    ssd.append(model.inertia_)
```

```python
plt.plot(range(2,30), ssd,'o--')
plt.xlabel('K Value')
plt.ylabel('ssd')
```

# Model Interpretation

```python
#let us choose k=15 as the drop in ssd value is low after k=15
```

```python
model = KMeans(n_clusters=15)
model.fit(scaled_X)
```

```python
model.labels_
```

```python
# inserting a new columns
X['K=15 Clusters'] = model.labels_
```
